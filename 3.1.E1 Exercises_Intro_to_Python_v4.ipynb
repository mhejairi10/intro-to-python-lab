#### __Variable:__
A variable is a labeled container that is used to store value(s). Look at it as a box where you label it with a name and store your item(s) in it.

**What is the purpose of a variable?**

It is to store, reference, and manipulate information.

**Naming Variables:**
* A variable name must start with a letter or the underscore character
* A variable name cannot start with a number
* A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
* Variable names are case-sensitive (number, Number, and NUMBER are three different variables)
* A variable name cannot be any of the Python keywords.

````python
# >> Instructions: Follow the comments step-by-step
# We will create a variable named 'score' and assign it a value of '90'. Note: to assign a value we use '='

score = 90

# To display a variable we use the 'print' function. This is an example of referencing a variable
# print the score variable

print(score)


# To manipulate a variable, Add 1 to the score value and update it. 
# Score should be 91.

score =score+1

print(score)
````

#### List methods

````python
# We have a list of fruits to buy
fruits = ["orange"]
# We need to add 'kiwi' to our list. Using append()

fruits.append("kiwi")

# fruits is now: ["orange", "kiwi"]
print(fruits)
````

````python
# Maybe we should get 'lemon' before 'kiwi' . Using insert()
fruits.insert(1,"Lemon")

# fruits is now: ["orange", "lemon", "kiwi"]
print(fruits)
````

````python
# Remove the 'kiwi' from our list. Using pop()
fruits.pop()
# fruits is now: ["orange", "lemon"]
print(fruits)

# Remove 'orange' too using pop() pointing to the index of it

fruits.pop(0)
# fruits is now: ["lemon"]
print(fruits)
````
